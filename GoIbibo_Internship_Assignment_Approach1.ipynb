### READING THE CONTENTS OF THE FILE

I have read the contents of the file by creating the handle and then calling the read method.

```python
f=open("drive/big.txt")
raw_text=f.read()
```

```python
words=raw_text.split()  
print(len(words)) 
```

Note that no of total words is 12773.

```python
print(len(set(words))) #  3687 unique words and 12773 total words in file.
```

Note that no of total unique words is 3687. This is often called as the **Type to Token Ratio (TTR).**

#### Now for now i am simply creating the dictionary where Key is the word itself and Value is the count or frequency. 

****We can also use the Count Vectorizer from scikit. I have shown this approach in the second notebook.****

```python
word_count_dict=dict()
for word in words:
  if word.lower() in word_count_dict:
        word_count_dict[word.lower()] += 1
  else:
        word_count_dict[word] = 1
```

```python
# word_count_dict
```

Now I am sorting this dictionary on the basis of the count values.

```python
sorted_word_dict_count={k: v for k, v in sorted(word_count_dict.items(), key=lambda item:item[1], reverse=True)}
```

```python
sorted_word_dict_count  
```

#### This dictionary has the required structure. 

****The key is the word and the value is the count of that word in the file. Also the words are arranged in non increasing order of count.****
